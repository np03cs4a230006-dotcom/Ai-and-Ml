## Section 4.1 — Task 1: Unit Converter

```python
def convert_length(value, direction):
    """
    Converts length between meters and feet.
    Parameters:
        value (float): The numeric value to convert.
        direction (str): '1' for meters->feet, '2' for feet->meters.
    Returns:
        float: The converted value.
    """
    if direction == '1':
        result = value * 3.28084
        print(f"{value} meters = {result:.4f} feet")
    else:
        result = value / 3.28084
        print(f"{value} feet = {result:.4f} meters")
    return result

def convert_weight(value, direction):
    """
    Converts weight between kilograms and pounds.
    Parameters:
        value (float): The numeric value to convert.
        direction (str): '1' for kg->lbs, '2' for lbs->kg.
    Returns:
        float: The converted value.
    """
    if direction == '1':
        result = value * 2.20462
        print(f"{value} kg = {result:.4f} lbs")
    else:
        result = value / 2.20462
        print(f"{value} lbs = {result:.4f} kg")
    return result

def convert_volume(value, direction):
    """
    Converts volume between liters and gallons.
    Parameters:
        value (float): The numeric value to convert.
        direction (str): '1' for liters->gallons, '2' for gallons->liters.
    Returns:
        float: The converted value.
    """
    if direction == '1':
        result = value * 0.264172
        print(f"{value} L = {result:.4f} gal")
    else:
        result = value / 0.264172
        print(f"{value} gal = {result:.4f} L")
    return result

def unit_converter():
    """
    Prompts the user to select a conversion type and value,
    performs the conversion and displays the result.
    Returns:
        float or None: Converted value, or None on invalid input.
    """
    print("\n=== Unit Converter ===")
    print("Choose conversion category:")
    print("1. Length  (meters <-> feet)")
    print("2. Weight  (kg <-> lbs)")
    print("3. Volume  (liters <-> gallons)")
    category = input("Enter category (1/2/3): ").strip()
    if category not in ('1', '2', '3'):
        print("Invalid category.")
        return None
    direction_labels = {
        '1': ("meters -> feet", "feet -> meters"),
        '2': ("kg -> lbs",      "lbs -> kg"),
        '3': ("liters -> gal",  "gal -> liters"),
    }
    labels = direction_labels[category]
    print(f"\nChoose direction:\n1. {labels[0]}\n2. {labels[1]}")
    direction = input("Enter direction (1/2): ").strip()
    if direction not in ('1', '2'):
        print("Invalid direction.")
        return None
    try:
        value = float(input("Enter the value to convert: "))
    except ValueError:
        print("Invalid input: please enter a numeric value.")
        return None
    converters = {'1': convert_length, '2': convert_weight, '3': convert_volume}
    return converters[category](value, direction)

unit_converter()
```

## Section 4.1 — Task 2: Math Operations on a List

```python
def compute_sum(numbers):
    """
    Returns the sum of a list of numbers.
    Parameters:
        numbers (list of float): Input numbers.
    Returns:
        float: Sum of all elements.
    """
    if not numbers:
        raise ValueError("The list is empty.")
    total = 0
    for n in numbers:
        total += n
    return total

def compute_average(numbers):
    """
    Returns the average of a list of numbers.
    Parameters:
        numbers (list of float): Input numbers.
    Returns:
        float: Mean of all elements.
    """
    if not numbers:
        raise ValueError("The list is empty.")
    return compute_sum(numbers) / len(numbers)

def compute_maximum(numbers):
    """
    Returns the maximum value in a list of numbers.
    Parameters:
        numbers (list of float): Input numbers.
    Returns:
        float: The largest element.
    """
    if not numbers:
        raise ValueError("The list is empty.")
    max_val = numbers[0]
    for n in numbers[1:]:
        if n > max_val:
            max_val = n
    return max_val

def compute_minimum(numbers):
    """
    Returns the minimum value in a list of numbers.
    Parameters:
        numbers (list of float): Input numbers.
    Returns:
        float: The smallest element.
    """
    if not numbers:
        raise ValueError("The list is empty.")
    min_val = numbers[0]
    for n in numbers[1:]:
        if n < min_val:
            min_val = n
    return min_val

def math_operations():
    """
    Prompts the user to select an operation and provide a list of numbers,
    then computes and displays the result.
    Returns:
        float or None: Result of the operation, or None on invalid input.
    """
    print("\n=== Math Operations ===")
    print("1. Sum\n2. Average\n3. Maximum\n4. Minimum")
    choice = input("Choose operation (1-4): ").strip()
    if choice not in ('1', '2', '3', '4'):
        print("Invalid choice.")
        return None
    raw = input("Enter numbers separated by spaces: ")
    try:
        numbers = [float(x) for x in raw.split()]
        if not numbers:
            raise ValueError("No numbers provided.")
    except ValueError as e:
        print(f"Invalid input: {e}")
        return None
    ops = {'1': compute_sum, '2': compute_average,
           '3': compute_maximum, '4': compute_minimum}
    try:
        result = ops[choice](numbers)
        print(f"Result: {result}")
        return result
    except ValueError as e:
        print(f"Error: {e}")
        return None

math_operations()
```

## Section 4.2 — List Manipulation

```python
# 1. Extract every other element
def extract_every_other(lst):
    """Returns every other element starting from index 0."""
    return lst[::2]

raw = input("1. Enter numbers separated by spaces: ")
lst = [int(x) for x in raw.split()]
print("Every other element:", extract_every_other(lst))
```

```python
# 2. Slice a sublist
def get_sublist(lst, start, end):
    """Returns elements from index start to end (inclusive)."""
    return lst[start:end + 1]

raw   = input("2. Enter numbers separated by spaces: ")
lst   = [int(x) for x in raw.split()]
start = int(input("   Enter start index: "))
end   = int(input("   Enter end index: "))
print("Sublist:", get_sublist(lst, start, end))
```

```python
# 3. Reverse a list using slicing
def reverse_list(lst):
    """Returns the list in reversed order using slicing."""
    return lst[::-1]

raw = input("3. Enter numbers separated by spaces: ")
lst = [int(x) for x in raw.split()]
print("Reversed list:", reverse_list(lst))
```

```python
# 4. Remove first and last elements
def remove_first_last(lst):
    """Returns the list without its first and last elements."""
    return lst[1:-1]

raw = input("4. Enter numbers separated by spaces: ")
lst = [int(x) for x in raw.split()]
print("After removing first and last:", remove_first_last(lst))
```

```python
# 5. Get first n elements
def get_first_n(lst, n):
    """Returns the first n elements of the list."""
    return lst[:n]

raw = input("5. Enter numbers separated by spaces: ")
lst = [int(x) for x in raw.split()]
n   = int(input("   Enter n: "))
print(f"First {n} elements:", get_first_n(lst, n))
```

```python
# 6. Extract last n elements
def get_last_n(lst, n):
    """Returns the last n elements of the list."""
    return lst[-n:]

raw = input("6. Enter numbers separated by spaces: ")
lst = [int(x) for x in raw.split()]
n   = int(input("   Enter n: "))
print(f"Last {n} elements:", get_last_n(lst, n))
```

```python
# 7. Reverse skip
def reverse_skip(lst):
    """Returns every second element starting from the second-to-last, moving backward."""
    return lst[-2::-2]

raw = input("7. Enter numbers separated by spaces: ")
lst = [int(x) for x in raw.split()]
print("Reverse skip result:", reverse_skip(lst))
```

## Section 4.3 — Nested Lists

```python
# 1. Flatten a nested list
def flatten(lst):
    """Flattens a one-level nested list into a single list."""
    result = []
    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result

print("1. Enter sublists one per line (e.g. 1 2 3). Blank line to finish:")
nested = []
while True:
    row = input()
    if row.strip() == '':
        break
    nested.append([int(x) for x in row.split()])
print("Flattened:", flatten(nested))
```

```python
# 2. Access nested element by indices
def access_nested_element(lst, indices):
    """Returns the element at the position described by the indices list."""
    element = lst
    for idx in indices:
        element = element[idx]
    return element

print("2. Enter sublists one per line. Blank line to finish:")
nested = []
while True:
    row = input()
    if row.strip() == '':
        break
    nested.append([int(x) for x in row.split()])
raw_idx = input("Enter indices separated by spaces (e.g. 1 2): ")
indices = [int(x) for x in raw_idx.split()]
print("Element at", indices, ":", access_nested_element(nested, indices))
```

```python
4# 3. Sum of all elements
def sum_nested(lst):
    """Recursively sums all numeric elements in a nested list."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

print("3. Enter sublists one per line. Blank line to finish:")
nested = []
while True:
    row = input()
    if row.strip() == '':
        break
    nested.append([int(x) for x in row.split()])
print("Sum of all elements:", sum_nested(nested))
```

```python
# 4. Remove specific element
def remove_element(lst, elem):
    """Removes all occurrences of elem from each sub-list."""
    return [[x for x in sublist if x != elem] for sublist in lst]

print("4. Enter sublists one per line. Blank line to finish:")
nested = []
while True:
    row = input()
    if row.strip() == '':
        break
    nested.append([int(x) for x in row.split()])
elem = int(input("Enter element to remove: "))
print("After removal:", remove_element(nested, elem))
```

```python
# 5. Find maximum element
def find_max(lst):
    """Recursively finds the maximum element in a nested list."""
    max_val = None
    for item in lst:
        candidate = find_max(item) if isinstance(item, list) else item
        if max_val is None or candidate > max_val:
            max_val = candidate
    return max_val

print("5. Enter sublists one per line. Blank line to finish:")
nested = []
while True:
    row = input()
    if row.strip() == '':
        break
    nested.append([int(x) for x in row.split()])
print("Maximum element:", find_max(nested))
```

```python
# 6. Count occurrences
def count_occurrences(lst, elem):
    """Recursively counts occurrences of elem in a nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

print("6. Enter sublists one per line. Blank line to finish:")
nested = []
while True:
    row = input()
    if row.strip() == '':
        break
    nested.append([int(x) for x in row.split()])
elem = int(input("Enter element to count: "))
print(f"'{elem}' appears", count_occurrences(nested, elem), "time(s)")
```

```python
# 7. Deep flatten
def deep_flatten(lst):
    """Recursively flattens a deeply nested list into a single flat list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print("7. Enter sublists one per line. Blank line to finish:")
nested = []
while True:
    row = input()
    if row.strip() == '':
        break
    nested.append([int(x) for x in row.split()])
print("Deep flattened:", deep_flatten(nested))
```

```python
# 8. Average of nested list
def average_nested(lst):
    """Returns the average of all numeric elements in a nested list."""
    all_elements = deep_flatten(lst)
    return sum(all_elements) / len(all_elements)

print("8. Enter sublists one per line. Blank line to finish:")
nested = []
while True:
    row = input()
    if row.strip() == '':
        break
    nested.append([int(x) for x in row.split()])
print("Average of all elements:", average_nested(nested))
```

## Section 10.1 — Problem 4: Matrix Operations

```python
import numpy as np

A = np.array([[3, 4], [7, 8]], dtype=float)
B = np.array([[5, 3], [2, 1]], dtype=float)

print("1. A * A_inv (should be Identity):\n",
      np.round(np.dot(A, np.linalg.inv(A)), 10))

AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2a. AB =\n", AB)
print("2b. BA =\n", BA)
print("   AB == BA?", np.allclose(AB, BA))

print("\n3a. (AB)^T =\n", AB.T)
print("3b. B^T * A^T =\n", np.dot(B.T, A.T))
print("   Equal?", np.allclose(AB.T, np.dot(B.T, A.T)))

coeff    = np.array([[ 2, -3,  1],
                     [ 1, -1,  2],
                     [ 3,  1, -1]], dtype=float)
constants = np.array([-1, -3, 9], dtype=float)

print("\n--- Solving:  2x-3y+z=-1 | x-y+2z=-3 | 3x+y-z=9 ---")
X_inv   = np.dot(np.linalg.inv(coeff), constants)
X_solve = np.linalg.solve(coeff, constants)
print("Method 1 (A_inv * B):    x={:.4f}, y={:.4f}, z={:.4f}".format(*X_inv))
print("Method 2 (linalg.solve): x={:.4f}, y={:.4f}, z={:.4f}".format(*X_solve))
print("Verification AX =>", np.round(np.dot(coeff, X_solve), 6))
```

## Section 10.2 — How Fast is NumPy?

```python
import numpy as np
import time

SIZE = 1_000_000

def speedup_msg(loop_t, np_t):
    print(f"  Python loop : {loop_t:.5f}s")
    print(f"  NumPy       : {np_t:.5f}s")
    print(f"  Speedup     : {loop_t / np_t:.1f}x faster\n")

# 1. Element-wise Addition
print("=" * 50)
print("1. Element-wise Addition (size 1,000,000)")
a_list = list(range(SIZE)); b_list = list(range(SIZE))
t0 = time.time()
_ = [a_list[i] + b_list[i] for i in range(SIZE)]
loop_add = time.time() - t0
a_np = np.array(a_list); b_np = np.array(b_list)
t0 = time.time()
_ = a_np + b_np
speedup_msg(loop_add, time.time() - t0)

# 2. Element-wise Multiplication
print("2. Element-wise Multiplication (size 1,000,000)")
t0 = time.time()
_ = [a_list[i] * b_list[i] for i in range(SIZE)]
loop_mul = time.time() - t0
t0 = time.time()
_ = a_np * b_np
speedup_msg(loop_mul, time.time() - t0)

# 3. Dot Product
print("3. Dot Product (size 1,000,000)")
t0 = time.time()
_ = sum(a_list[i] * b_list[i] for i in range(SIZE))
loop_dot = time.time() - t0
t0 = time.time()
_ = np.dot(a_np, b_np)
speedup_msg(loop_dot, time.time() - t0)

# 4. Matrix Multiplication 1000x1000
print("4. Matrix Multiplication (1000x1000)")
MAT = 1000
m1 = [[float(i*MAT+j) for j in range(MAT)] for i in range(MAT)]
m2 = [[float(i*MAT+j) for j in range(MAT)] for i in range(MAT)]
t0 = time.time()
_ = [[sum(m1[i][k]*m2[k][j] for k in range(MAT)) for j in range(MAT)] for i in range(MAT)]
loop_mat = time.time() - t0
m1_np = np.array(m1); m2_np = np.array(m2)
t0 = time.time()
_ = np.dot(m1_np, m2_np)
speedup_msg(loop_mat, time.time() - t0)
```
